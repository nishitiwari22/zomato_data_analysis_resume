```python
df = pd.read_csv("../data/clean/zomato_cleaned.csv")
df.head()
```

⭐ This is the KEY step

Define:

High Rated → rating ≥ 4.0 → 1

Low Rated → rating < 4.0 → 0

```python
df['rate'].unique()[:10]
```

```python
# Remove '/5'
df['rate'] = df['rate'].str.replace('/5', '', regex=False)

# Replace invalid ratings with NaN
df['rate'] = df['rate'].replace(['NEW', '-'], np.nan)

# Convert to float
df['rate'] = df['rate'].astype(float)
```

```python
df['rate'].dtype
```

```python
df = df.dropna(subset=['rate'])
```

```python
df['rating_class'] = np.where(df['rate'] >= 4.0, 1, 0)
```

```python
df[['rate', 'rating_class']].head()
```
